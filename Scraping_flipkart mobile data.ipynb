```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
Product_name = []
Price = []
Image = []
```

```python
for i in range(1,42):
    
    URL = f"https://www.flipkart.com/clothing-and-accessories/pr?sid=clo&otracker=categorytree&p%5B%5D=facets.ideal_for%255B%255D%3DMen&otracker=nmenu_sub_Men_0_Clothing&page={i}"

    HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
    }

    
    page = requests.get(URL, headers=HEADERS)

    soup = BeautifulSoup(page.content, "html.parser")

    box = soup.find("div",class_="DOjaWF YJG4Cf")

    if box is not None:
        product_container = box.find_all("div", class_="_1sdMkc LFEi7Z")

        for container in product_container:
            image_tag = container.find("a", class_="rPDeLR")
            image = image_tag.find("img") if image_tag else None
            Image.append(image['src'] if image else "")

            name = container.find("div", class_="syl9yP")
            Product_name.append(name.text.strip() if name else "")

            pricing = container.find("div", class_="hl05eU")
            Price.append(pricing.text.strip() if pricing else "")
    else:
        print(f"No product container found on page {i}")

for name, price, image in zip(Product_name, Price, Image):
    print(f"{name} | {price} | {image}")
```

```python
df = pd.DataFrame({
    "Product Name":Product_name,
    "Prices":Price,
    "Image":Image
})
```

```python
df.to_csv('Webscrapping.csv')
```
