```python
import pandas as pd
```

```python
df = pd.read_csv('Webscrapping.csv')
```

```python
df
```

```python
df.drop(columns='Unnamed: 0',inplace=True)
```

```python
df['Prices'] = df['Prices'].str.split('₹').str[1]
```

```python
df['Product Name'] = df['Product Name'].str.split(" ").str[0]
```

```python
df.head()
```
